#2.2 Two Categorical Variables and Cross-Tabulations

To investigate the relationship between variables, we can look at the marginal and join distributions. 

```python
pd.crosstab(df_flying_etiquette["Do you ever recline your seat when you fly?"], df_flying_etiquette["How tall are you?"])
```

```python
joint_recline_height = pd.crosstab(df_flying_etiquette["Do you ever recline your seat when you fly?"], df_flying_etiquette["How tall are you?"], 
                                    normalize=True)
joint_recline_height
```

```python
height_marginal = joint_recline_height.sum(axis=0)
recline_marginal = joint_recline_height.sum(axis=1)
```

```python
pd.crosstab(df_flying_etiquette["Do you ever recline your seat when you fly?"], df_flying_etiquette["How tall are you?"], 
                                    normalize=True, margins=True)
```

```python
cond_recline_height = joint_recline_height.divide(height_marginal, axis=1)
cond_recline_height 
```

```python
cond_recline_height.T.plot.bar(stacked=True)
```

#2.3 Multi-Way Tables and Simpson's Paradox


We can add a third variable into our analysis and see how it affects the frequencies. 

```python
joint_recline_height_travelfreq = pd.crosstab(
    [df_flying_etiquette["How tall are you?"], df_flying_etiquette["How often do you travel by plane?"]],
    df_flying_etiquette["Do you ever recline your seat when you fly?"],
    normalize=True
)
joint_recline_height_travelfreq
```

```python
joint_height_travelfreq = joint_recline_height_travelfreq.sum(axis=1)
```

```python
recline_given_height_travelfreq = joint_recline_height_travelfreq.divide(
    joint_height_travelfreq,
    axis=0
)
recline_given_height_travelfreq
```
