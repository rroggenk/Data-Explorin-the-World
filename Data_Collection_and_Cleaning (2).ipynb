#Gapminder Data
https://www.gapminder.org/data/

The data from Gapminder is simply to add some more variables from the international tourism data for each country in each year, so the observational units are the same as the international tourism data. 

CO2 emissions (tonnes per person)

```python
df_co2_emissions = pd.read_csv(data_dir + "co2_emissions_tonnes_per_person.csv")
df_co2_emissions = df_co2_emissions.rename(columns={"country": "Country Name"})
df_co2_emissions.head()
```

```python
years = []
for i in range(1799, 2018):
  years.append(str(i))

df_co2_emissions_long = pd.melt(df_co2_emissions, id_vars = "Country Name", value_vars = years, var_name = "year", value_name = "co2_emissions")
df_co2_emissions_long = df_co2_emissions_long.set_index(["Country Name", "year"])
df_co2_emissions_long.head()
```

```python
df_co2_emissions_long = df_co2_emissions_long.replace("60µ", 0)
df_co2_emissions_long = df_co2_emissions_long.replace("70µ", 0)
df_co2_emissions_long = df_co2_emissions_long.replace("80µ", 0)
df_co2_emissions_long = df_co2_emissions_long.replace("90µ", 0)
df_co2_emissions_long["co2_emissions"] = pd.to_numeric(df_co2_emissions_long["co2_emissions"])
df_co2_emissions_long.head()
```

Income per person (GDP/capita, PPP$ inflation-adjusted)

```python
df_income_per_person = pd.read_csv(data_dir + "income_per_person_gdppercapita_ppp_inflation_adjusted.csv")
df_income_per_person = df_income_per_person.rename(columns={"country": "Country Name"})
df_income_per_person.head()
```

```python
years = []
for i in range(1799, 2050):
  years.append(str(i))

df_income_per_person_long = pd.melt(df_income_per_person, id_vars = "Country Name", value_vars = years, var_name = "year", value_name = "income_per_person")
df_income_per_person_long = df_income_per_person_long.set_index(["Country Name", "year"])
```

```python
df_income_per_person_long["income_per_person"]= df_income_per_person_long["income_per_person"].astype(str)

for i in range(len(df_income_per_person_long["income_per_person"])):
  val = df_income_per_person_long["income_per_person"][i]
  if "k" in val:
    val = val[:-1]
    val = float(val)
    val = val*1000
    df_income_per_person_long["income_per_person"][i] = val

df_income_per_person_long["income_per_person"] = pd.to_numeric(df_income_per_person_long["income_per_person"])
df_income_per_person_long.head()
```

Population

```python
df_population = pd.read_csv(data_dir + "population_total.csv")
df_population = df_population.rename(columns={"country": "Country Name"})
df_population.head()
```

```python
years = []
for i in range(1799, 2100):
  years.append(str(i))

df_population_long = pd.melt(df_population, id_vars = "Country Name", value_vars = years, var_name = "year", value_name = "population")
df_population_long = df_population_long.set_index(["Country Name", "year"])
```

```python
df_population_long["population"]= df_population_long["population"].astype(str)

for i in range(len(df_population_long["population"])):
  val = df_population_long["population"][i]
  if "k" in val:
    val = val[:-1]
    val = float(val)
    val = val*1000
    df_population_long["population"][i] = val
  elif "M" in val:
    val = val[:-1]
    val = float(val)
    val = val*1000000
    df_population_long["population"][i] = val
  elif "B" in val:
    val = val[:-1]
    val = float(val)
    val = val*1000000000
    df_population_long["population"][i] = val

df_population_long["population"] = pd.to_numeric(df_population_long["population"])
df_population_long.head()
```

Total GDP (US$, inflation-adjusted)

```python
df_total_gdp = pd.read_csv(data_dir + "total_gdp_us_inflation_adjusted.csv")
df_total_gdp = df_total_gdp.rename(columns={"country": "Country Name"})
df_total_gdp.head()
```

```python
years = []
for i in range(1959, 2020):
  years.append(str(i))

df_total_gdp_long = pd.melt(df_total_gdp, id_vars = "Country Name", value_vars = years, var_name = "year", value_name = "total_gdp")
df_total_gdp_long = df_total_gdp_long.set_index(["Country Name", "year"])
df_total_gdp_long.head()
```

```python
df_total_gdp_long["total_gdp"]= df_total_gdp_long["total_gdp"].astype(str)

for i in range(len(df_total_gdp_long["total_gdp"])):
  val = df_total_gdp_long["total_gdp"][i]
  if "k" in val:
    val = val[:-1]
    val = float(val)
    val = val*1000
    df_total_gdp_long["total_gdp"][i] = val
  elif "M" in val:
    val = val[:-1]
    val = float(val)
    val = val*1000000
    df_total_gdp_long["total_gdp"][i] = val
  elif "B" in val:
    val = val[:-1]
    val = float(val)
    val = val*1000000000
    df_total_gdp_long["total_gdp"][i] = val
  elif "TR" in val:
    val = val[:-2]
    val = float(val)
    val = val*1000000000000
    df_total_gdp_long["total_gdp"][i] = val

df_total_gdp_long["total_gdp"] = pd.to_numeric(df_total_gdp_long["total_gdp"], errors="coerce")
df_total_gdp_long.head()
```

#Country -> Continent Data
https://www.kaggle.com/datasets/statchaitya/country-to-continent?resource=download

The data from Kaggle is simply to have the continent and subregion of each country in our data set in case we want to broaden our visuals to group by continent or subregion instead of each country indiviually. 

```python
df_country_continent = pd.read_csv(data_dir + "countryContinent.csv", encoding="ISO-8859-1")
df_country_continent = df_country_continent[["country", "continent", "sub_region"]]
df_country_continent = df_country_continent.rename(columns={"country": "Country Name"})
df_country_continent.head()
```

#Merging All Datasets and Saving International Tourism Dataset as a CSV File

```python
df_international_tourism = df_expenditures_percent_total_imports_long.merge(df_expenditures_current_us_long, on=["Country Name", "year"])
df_international_tourism = df_expenditures_passenger_transport_items_long.merge(df_international_tourism, on=["Country Name", "year"])
df_international_tourism = df_number_of_arrivals_long.merge(df_international_tourism, on=["Country Name", "year"])
df_international_tourism = df_number_of_departures_long.merge(df_international_tourism, on=["Country Name", "year"])
df_international_tourism = df_receipts_current_us_long.merge(df_international_tourism, on=["Country Name", "year"])
df_international_tourism = df_receipts_passenger_transport_items_long.merge(df_international_tourism, on=["Country Name", "year"])
df_international_tourism = df_receipts_travel_items_long.merge(df_international_tourism, on=["Country Name", "year"])
df_international_tourism = df_receipts_percent_total_exports_long.merge(df_international_tourism, on=["Country Name", "year"])

df_international_tourism = df_co2_emissions_long.merge(df_international_tourism, on=["Country Name", "year"])
df_international_tourism = df_income_per_person_long.merge(df_international_tourism, on=["Country Name", "year"])
df_international_tourism = df_population_long.merge(df_international_tourism, on=["Country Name", "year"])
df_international_tourism = df_total_gdp_long.merge(df_international_tourism, on=["Country Name", "year"])

df_international_tourism = df_international_tourism.reset_index()

df_international_tourism = df_international_tourism.merge(df_country_continent, on="Country Name", how="left")

df_international_tourism
```
